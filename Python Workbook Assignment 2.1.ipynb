Import the Libraries

```python
import pandas as pd
import numpy as np
import glob
```

Create an Empty DataFrame

```python
all_data=pd.DataFrame()
```

Loop through all the 104 files matching the pattern and load the excel file into the DataFrame and then appending the contents to the DataFrame created above

```python
for f in glob.glob("datasets/datasets/weekly_call_data/weekdata_*.xlsx"):
    df=pd.read_excel(f)
    print(f)
    all_data=all_data.append(df, ignore_index=True)
```
